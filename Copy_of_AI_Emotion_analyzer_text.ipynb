```python
import spacy
import en_core_web_lg
from spacy.lookups import Lookups
from spacy.training import Example
from spacy.util import minibatch, compounding
import pandas as pd
import random
from sklearn.model_selection import train_test_split

nlp = spacy.blank("en")

# Add textcat pipe without any config (default settings)
textcat = nlp.add_pipe("textcat")
balanced_df = (
    df9.groupby("label")
    .apply(lambda x: x.sample(n=min(1500, len(x)), random_state=42))
    .reset_index(drop=True)
)
labels = [str(l) for l in sorted(balanced_df["label"].unique())]
# Add your labels (as strings)
labels = [str(l) for l in sorted(balanced_df["label"].unique())]

# Sample dataset (replace with your data)


# Split data
train_df, test_df = train_test_split(balanced_df, test_size=0.2, stratify=balanced_df["label"])



# Convert dataframe rows to spaCy examples
def convert_to_examples(nlp, dataframe):
    examples = []
    for text, label in zip(dataframe["content"], dataframe["label"]):
        cats = {l: 0.0 for l in labels}
        cats[str(label)] = 1.0
        doc = nlp.make_doc(text)
        examples.append(Example.from_dict(doc, {"cats": cats}))
    return examples

train_examples = convert_to_examples(nlp, train_df)
test_examples = convert_to_examples(nlp, test_df)

# Initialize pipeline with training data
nlp.initialize(get_examples=lambda: train_examples)

# Train loop
n_iter = 5
for i in range(n_iter):
    random.shuffle(train_examples)
    losses = {}
    batches = minibatch(train_examples, size=compounding(4.0, 16.0, 1.001))
    for batch in batches:
        nlp.update(batch, losses=losses)
    print(f"Epoch {i+1} Loss: {losses.get('textcat', 0):.4f}")

# Evaluate accuracy

correct = 0
for example in test_examples:
    doc = nlp(example.text)
    pred_label = max(doc.cats, key=doc.cats.get)
    true_label = max(example.reference.cats, key=example.reference.cats.get)
    print(f"Text: {example.text}")
    print(f"Predicted: {repr(pred_label)} ({type(pred_label)}), True: {repr(true_label)} ({type(true_label)})")
    if str(pred_label) == str(true_label):  # compare as strings
        correct += 1

accuracy = correct / len(test_examples)
print(f"Accuracy: {accuracy:.4f}")

```

```python


# Dictionary mapping emotion words to numbers
word_to_number = {
    "sadness": 0,
    "approval": 9,
    "amusement": 7,
    "annoyance": 8,
    "surprise": 5,
    "love": 2,
    "caring": 10,
    "curiosity": 12,
    "confusion": 11,
    "admiration": 6,
    "anger": 3,
    "fear": 4,
    "disappointment": 13,
    "disapproval": 14,
    "disgust": 15,
    "embarrassment": 16,
    "excitement": 17,
    "gratitude": 18,
    "joy": 1,
    "optimism": 19,
    "realization": 20,
    "remorse": 21,
    "neutral": 22,
}

# Reverse it properly: keys as strings of numbers
number_to_word = {str(v): k for k, v in word_to_number.items()}

def predict_emotion(text):
    doc = nlp(text)

    # Get top prediction
    top_label = max(doc.cats, key=doc.cats.get)
    top_confidence = doc.cats[top_label]

    # Translate number label to emotion word
    emotion_name = number_to_word.get(top_label, f"Unknown ({top_label})")

    # Sort all emotions
    sorted_emotions = sorted(doc.cats.items(), key=lambda x: x[1], reverse=True)

    # Format result
    result = f"**Top Emotion:** `{emotion_name}` ({top_confidence:.2f})\n\n"
    result += "**All Predictions:**\n"
    for label, score in sorted_emotions:
        emotion = number_to_word.get(label, f"Unknown ({label})")
        result += f"- {emotion}: {score:.2f}\n"

    return result

# Gradio interface
iface = gr.Interface(
    fn=predict_emotion,
    inputs=gr.Textbox(lines=2, placeholder="Type a sentence here..."),
    outputs="markdown",
    title="🧠 Text Emotion Classifier",
    description="This AI detects the emotional tone in your sentence.",
    allow_flagging="never"
)

iface.launch()

# Gradio UI setup
iface = gr.Interface(
    fn=predict_emotion,
    inputs=gr.Textbox(lines=2, placeholder="Enter a sentence..."),
    outputs="markdown",
    title="🧠 Emotion Predictor",
    description="Enter a sentence to see which emotion it most likely represents.",
    allow_flagging="never"
)

# Launch the app
iface.launch()

```

```python
number_to_word = {
    0: "sadness",
    1: "joy",
    2: "love",
    3: "anger",
    4: "fear",
    5: "surprise",
    6: "admiration",
    7: "amusement",
    8: "annoyance",
    9: "approval",
    10: "caring",
    11: "confusion",
    12: "curiosity",
    13: "disappointment",
    14: "disapproval",
    15: "disgust",
    16: "embarrassment",
    17: "excitement",
    18: "gratitude",
    19: "optimism",
    20: "realization",
    21: "remorse",
    22: "neutral"
}

def predict_emotion(text):
    doc = nlp(text)

    # Get top prediction
    top_label = max(doc.cats, key=doc.cats.get)
    top_confidence = doc.cats[top_label]

    # Convert number to emotion
    try:
        emotion_name = number_to_word[int(top_label)]
    except:
        emotion_name = "Unknown"

    result = f"**Top Emotion:** `{emotion_name}` (score: {top_confidence:.2f})\n\n"
    result += "**All Predictions:**\n"

    for k, v in sorted(doc.cats.items(), key=lambda x: x[1], reverse=True):
        try:
            name = number_to_word[int(k)]
        except:
            name = "Unknown"
        result += f"- {name} ({k}): {v:.2f}\n"

    return result

# Gradio Interface
iface = gr.Interface(
    fn=predict_emotion,
    inputs=gr.Textbox(lines=2, placeholder="Type your sentence..."),
    outputs="markdown",
    title="Emotion Detector",
    description="Detect emotions from text using AI",
    allow_flagging="never"
)

iface.launch()
```
